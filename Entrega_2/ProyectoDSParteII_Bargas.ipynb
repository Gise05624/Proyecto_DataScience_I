## üìä An√°lisis de datos exploratorios (EDA)
A continuaci√≥n, comparto algunas visualizaciones que me ayudaron a entender c√≥mo se comportan las variables m√°s importantes del an√°lisis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('ventas_dataset_actdeclase.csv')
```

```python
df = pd.read_csv('ventas_dataset_actdeclase.csv')
df['Mes'] = pd.to_datetime(df['Mes'])
```

```python
ventas_mensuales = df.groupby(['Mes', 'Region'])['Ventas'].sum().unstack()
ventas_mensuales.plot(marker='o', figsize=(10, 6))
plt.title("Ventas mensuales por regi√≥n")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.grid(True)
plt.savefig("ventas_mensuales.png")
plt.close()
```

```python
from IPython.display import Image
Image(filename="ventas_mensuales.png")
```

### Boxplot Satisfaccion


```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Region', y='Satisfaccion_Cliente')
plt.title("Distribuci√≥n de satisfacci√≥n del cliente por regi√≥n")
plt.savefig("boxplot_satisfaccion.png")
plt.close()
```

```python
Image(filename='boxplot_satisfaccion.png')
```

### Correlacion


```python
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Ventas', 'Satisfaccion_Cliente', 'Costo_Operativo']].corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de correlaci√≥n")
plt.savefig("correlacion.png")
plt.close()
```

```python
Image(filename='correlacion.png')
```

### Scatter Satisfaccion Ventas


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Satisfaccion_Cliente', y='Ventas', hue='Region')
plt.title("Relaci√≥n entre satisfacci√≥n del cliente y ventas")
plt.savefig("scatter_satisfaccion_ventas.png")
plt.close()
```

```python
Image(filename='scatter_satisfaccion_ventas.png')
```

### Scatter Costos Ventas


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Costo_Operativo', y='Ventas', hue='Region')
plt.title("Relaci√≥n entre costos operativos y ventas")
plt.savefig("scatter_costos_ventas.png")
plt.close()
```

## üìë Tipo de Aprendizaje y Propuesta de Modelos

Este caso encaja dentro de un aprendizaje supervisado, porque se podr√≠a intentar predecir una variable objetivo (por ejemplo, ventas) en base a otras variables como costos y satisfacci√≥n.

Modelos posibles:
* Regresi√≥n Lineal o √Årboles de Decisi√≥n ‚Üí para predecir el nivel de ventas.

* Modelos de Clasificaci√≥n ‚Üí si se quiere agrupar las regiones seg√∫n rendimiento (alta, media o baja).

* Clustering (no supervisado) ‚Üí para encontrar grupos de productos o regiones con comportamientos parecidos.

Estos modelos permitir√≠an anticiparse a ciertos resultados o mejorar la segmentaci√≥n dentro del negocio.
